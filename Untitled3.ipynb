```python
# Step 1: Import Libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Step 2: Load the Dataset
url = "https://raw.githubusercontent.com/CSSEGISandData/COVID-19/master/csse_covid_19_data/csse_covid_19_time_series/time_series_covid19_confirmed_global.csv"
confirmed_cases = pd.read_csv(url)

# Display the first few rows of the dataset
print(confirmed_cases.head())

# Step 3: Data Cleaning
# Drop unnecessary columns (e.g., Lat, Long)
confirmed_cases = confirmed_cases.drop(columns=["Lat", "Long"])

# Aggregate data by country
confirmed_cases = confirmed_cases.groupby("Country/Region").sum()

# Transpose the data for easier analysis (rows = dates, columns = countries)
confirmed_cases = confirmed_cases.T

# Extract the date columns (skip the first row which contains 'Province/State', 'Country/Region', etc.)
dates = confirmed_cases.index[4:]  # Skip the first 4 rows (non-date rows)
confirmed_cases = confirmed_cases.iloc[4:]  # Keep only rows with dates

# Convert the index to datetime format
confirmed_cases.index = pd.to_datetime(confirmed_cases.index, format="%m/%d/%y")

# Display the cleaned dataset
print(confirmed_cases.head())

# Step 4: Exploratory Data Analysis (EDA)
# Total confirmed cases globally over time
global_cases = confirmed_cases.sum(axis=1)

# Plot global cases over time
plt.figure(figsize=(10, 6))
plt.plot(global_cases.index, global_cases, label="Global Confirmed Cases", color="blue")
plt.title("Global COVID-19 Confirmed Cases Over Time")
plt.xlabel("Date")
plt.ylabel("Number of Cases")
plt.legend()
plt.grid()
plt.show()

# Top 10 countries with the highest number of cases
top_10_countries = confirmed_cases.iloc[-1].sort_values(ascending=False).head(10)

# Plot top 10 countries
plt.figure(figsize=(10, 6))
#sns.barplot(x=top_10_countries.values, y=top_10_countries.index, palette="viridis")
sns.barplot(x=top_10_countries.values, y=top_10_countries.index, hue=top_10_countries.index, palette="viridis", legend=False)

plt.title("Top 10 Countries with Highest COVID-19 Cases")
plt.xlabel("Number of Cases")
plt.ylabel("Country")
plt.show()

# Step 5: Analyze Specific Countries
# Compare trends for specific countries (e.g., US, India, Brazil)
countries = ["US", "India", "Brazil"]
plt.figure(figsize=(10, 6))
for country in countries:
    plt.plot(confirmed_cases.index, confirmed_cases[country], label=country)

plt.title("COVID-19 Cases in Selected Countries")
plt.xlabel("Date")
plt.ylabel("Number of Cases")
plt.legend()
plt.grid()
plt.show()

# Step 6: Insights and Reporting
# Print summary statistics
print("Global Confirmed Cases Summary:")
print(global_cases.describe())

# Print the latest global cases
#print(f"Latest Global Confirmed Cases: {global_cases[-1]}")
print(f"Latest Global Confirmed Cases: {global_cases.iloc[-1]}")
```
